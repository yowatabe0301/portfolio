# 3. Assumptions

In the process of forecasting the production of oil and gas wells, several assumptions have been made due to limited information or the constraints of the available dataset. Here are the key assumptions and the rationale behind them:

1. **Stable Production Factors**: It's assumed that external factors affecting production, such as market demand, geopolitical stability, and technological advancements, remain constant or predictable based on historical trends. This assumption is necessary because modeling these factors requires complex, multi-variable analysis beyond the scope of our current dataset.

2. **Data Representativeness**: We assume the data provided represents the broader population of oil and gas wells accurately. This includes well characteristics, production volumes, and operational parameters. The reason for this assumption is the reliance on historical data to predict future outcomes. If the data is not representative, the forecasts may not be accurate.

3. **Consistent Data Quality**: The analysis assumes that the historical production data and well characteristics are accurate, complete, and consistently measured across all entries. This is crucial for the reliability of our predictive models. In reality, discrepancies or errors in data collection methods could affect model performance, but addressing these issues falls outside the immediate analytical scope.

4. **Predictive Power of Historical Data**: Another assumption is that past well performance is indicative of future production. This underpins the entire forecasting model, relying on the principle that historical trends and patterns can inform future expectations. While generally a reasonable assumption, it may not hold if there are significant changes in extraction technologies, regulatory environments, or well depletion rates.

5. **No Sudden Regulatory Changes**: It's assumed there will be no sudden regulatory changes that could significantly impact well operations and production. Regulatory shifts can dramatically alter production capabilities and economic viability. This assumption allows us to focus on the data-driven aspects of production forecasting without delving into speculative policy analysis.

6. **Uniform Operational Practices**: We assume that operational practices and technologies across the wells in the dataset are relatively uniform or have a predictable impact on production. This simplifies the model by not requiring detailed data on operational nuances for each well, which may not be available.

7. **Data Completeness and Accuracy**: The analysis assumes that the provided datasets are comprehensive and accurately reflect the wells' historical performance, operational costs, and production metrics. This assumption underlies the model's development and is vital for ensuring that the forecasts and insights generated are based on a true representation of reality.

By stating these assumptions, we aim to clarify the basis on which the predictive models have been developed. It's important for stakeholders to understand these assumptions for informed decision-making, acknowledging the potential limitations they introduce to the forecasts and recommendations.

# 4. Limitations

1. **Data Limitations**: Our models are built on the historical production and geological data provided. Such external factors can introduce volatility that our models, based on historical data, cannot predict.

4. **Predictive Uncertainty**: While statistical models can capture trends and patterns within the data, there is inherent uncertainty in any forecast. This uncertainty increases with the forecast horizon. Our models' predictions should be viewed as one possible future scenario among many, rather than definitive forecasts.

5. **Model Assumptions**: The assumptions underlying the ARIMA and Exponential Smoothing models, such as stationarity or seasonality patterns, may not hold true across all wells or throughout the entire dataset. Violations of these assumptions could compromise model performance and forecast accuracy.

6. **Data-Driven Insights**: Our recommendations are derived from model outputs and data-driven insights. For a non-technical audience, an improvement in data dictionaries could have served as a basis for a more comprehensive understanding of the industry and its variables. If given more comprehensive and cleaner data overall, we anticipate that model accuracy and the robustness of our recommendations could be significantly improved. The current limitations highlight the importance of continuous data collection, model refinement, and the integration of additional data sources for enhancing predictive capabilities.

In light of these limitations, stakeholders should use the forecasts and recommendations provided as one component of a broader decision-making framework, considering other qualitative and quantitative factors to guide strategic and operational decisions in the oil and gas industry.

## Creating Clusters

Our decision to create three clusters for both oil and gas wells stem from our aim to enhance the accuracy of our forecasting by categorizing wells based on their characteristics. After analyzing correlations, we selected key geological attributes that are strongly associated with oil and gas production for clustering purposes. Utilizing the K-Means clustering algorithm, we segmented the wells into distinct clusters based on these attributes. This approach enables us to identify different types of wells with similar geological characteristics, allowing for more precise forecasting models tailored to each cluster's unique properties. Additionally, we sought to analyze the lifecycle of the wells based on their production start dates. By extracting the year from the first production date and grouping the wells accordingly, we categorized them into 'Old', 'Middle' and 'Young' stages, providing insights into the age distribution of the wells and aiding in strategic decision-making processes. Overall, these clustering and lifecycle analysis techniques enable us to gain deeper insights into well behavior and optimize our forecasting models for improved performance and accuracy. 

_Note: Life Cycle Stages: 'Old': Years 2009-2019, 'Middle': Years 2020-2021, 'Young': Years 2022-2023_

```python
# Normalize the relevant data for the Oil Cluster
features_to_cluster_oil = ['OilEURWH_MBBL', 'TopOfZone_FT', 'BottomOfZone_FT', 'EffectivePorosity_PCT', 'DensityPorosity_PCT',
                          'TotalOrganicCarbon_WTPCT']
scaler_oil = StandardScaler()  # Standardize features by removing the mean and scaling to unit variance
metadata_normalized_oil = scaler_oil.fit_transform(metadata[features_to_cluster_oil])  # Fit to data, then transform it

# Apply K-Means clustering for the Oil Cluster
kmeans_oil = KMeans(n_clusters=5, random_state=0)  # Instantiate KMeans with 5 clusters
metadata['Oil_Cluster'] = kmeans_oil.fit_predict(metadata_normalized_oil)  # Compute cluster centers and predict cluster index for each sample

# Normalize the relevant data for the Gas Cluster
features_to_cluster_gas = ['GasEURWH_BCF', 'WaterSaturation_PCT', 'ClayVolume_PCT']
scaler_gas = StandardScaler()  # Standardize features for gas cluster
metadata_normalized_gas = scaler_gas.fit_transform(metadata[features_to_cluster_gas])  # Fit to data, then transform it

# Apply K-Means clustering for the Gas Cluster
kmeans_gas = KMeans(n_clusters=5, random_state=0)  # Instantiate KMeans with 5 clusters
metadata['Gas_Cluster'] = kmeans_gas.fit_predict(metadata_normalized_gas)  # Compute cluster centers and predict cluster index for each sample

# Extract the year from 'FirstProdDate'
metadata['FirstProdYear'] = pd.to_datetime(metadata['FirstProdDate']).dt.year  # Extract year from datetime 'FirstProdDate'

# Manual assignment of 'Lifecycle_Stage' based on 'FirstProdYear'
def assign_custom_lifecycle_stage(year):
    if year <= 2017:  # Older fields before 2017
        return 'Old'
    elif 2017 < year <= 2020:  # Middle-aged fields between 2017 and 2020
        return 'Middle'
    else:  # Younger fields from 2021 onwards
        return 'Young'

# Apply the custom function to assign 'Lifecycle_Stage' based on 'FirstProdYear'
metadata['Lifecycle_Stage'] = metadata['FirstProdYear'].apply(assign_custom_lifecycle_stage)  # Apply custom function to 'FirstProdYear'

# Display the first few rows of the dataframe with new cluster labels and lifecycle stages
metadata[['FirstProdDate', 'TotalProducingMonths', 'PeakProductionDate', 'CumProd_BOE', 'Lifecycle_Stage', 'Oil_Cluster', 'Gas_Cluster']].head()
```

```python
# Group by 'Lifecycle_Stage' and calculate minimum and maximum production years for each cluster
cluster_year_ranges = metadata.groupby('Lifecycle_Stage')['FirstProdYear'].agg(['min', 'max'])

# Calculate the range of years for each cluster by subtracting minimum year from maximum year
cluster_year_ranges['Year_Range'] = cluster_year_ranges['max'] - cluster_year_ranges['min']

# Display the min, max, and range of years for each lifecycle stage
print("Year Ranges for each Lifecycle Stage Cluster:")
print(cluster_year_ranges)
```

The results of the lifecycle clustering reveal distinct year ranges for each lifecycle stage cluster: 'Old', 'Middle' and 'Young'. Interestingly, we observe that the "Old" cluster encompasses wells with production start dates ranging from 2009 to 2017, indicating a longer operational history compared to the other stages. Similarly, the 'Middle' cluster includes wells starting from 2018 to 2020, while the 'Young' cluster comprises wells with production start dates from 2021 to 2023. While the year ranges provide insights into the age distribution of the wells, it is worth noting that our initial intention was to achieve a more balanced distribution of wells across the clusters. However, due to certain predictive modeling requirements necessitating more than 12 months of data, wider year ranges were adopted. Consequently, this led to an unequal distribution of wells among the clusters.

## Total Producing Months & Clusters 

```python
# We are defined a function 'plot_production_by_cluster_and_metric' to visualize production trends by cluster type and chosen metric.
# This function aggregates production data using 'TotalProdMonths' as the unit of time.
def plot_production_by_cluster_and_metric(cluster_type, volume_metric):
    # Grouping the production data by cluster type and 'TotalProdMonths', then summing the chosen production volume metric
    production_by_cluster = production.groupby([cluster_type, 'ProducingMonth'])[volume_metric].sum().reset_index()

    # Plotting the production trends for each cluster type
    plt.figure(figsize=(15, 7))
    for label, grp in production_by_cluster.groupby(cluster_type):
        plt.plot(grp['ProducingMonth'], grp[volume_metric], marker='o', linestyle='-', label=f'{label}')
    
    plt.title(f'Total Production by {cluster_type} using {volume_metric}')
    plt.xlabel('Total Production Months')
    plt.ylabel(f'Production ({volume_metric})')
    plt.grid(True)
    plt.legend(title=cluster_type)
    plt.show()

# We have defined a list of cluster types and their corresponding production volume metrics.
# Each tuple represents a pair of cluster type and metric for plotting production trends.
cluster_types_and_metrics = [
    ('Oil_Cluster', 'LiquidsProd_BBL'),  # Production trends for oil clusters based on liquid production
    ('Gas_Cluster', 'GasProd_MCF'),  # Production trends for gas clusters based on gas production
    ('Lifecycle_Stage', 'GasProd_MCF'),  # Production trends for different lifecycle stages based on gas production
    ('Lifecycle_Stage', 'LiquidsProd_BBL')  # Production trends for different lifecycle stages based on liquid production
]

# We iterate through the list of cluster types and metrics to generate production plots for each combination.
for cluster_type, volume_metric in cluster_types_and_metrics:
    plot_production_by_cluster_and_metric(cluster_type, volume_metric)
```

In the first graph, 'Total Production by Oil Cluster using LiquidsProd_BBL', we observe multiple lines representing different clusters within oil production. Each line traces the production quantity measured in barrels over the total production months. From the initial period around 2010, the production for all clusters starts relatively low. Around 2014, there is a noticeable increase in production, with one cluster in particular (purple line) showing a significant surge, peaking between 2020 and 2022 before showing signs of decline. The other clusters follow similar upward trends, albeit to varying degrees and with different peak timings.

The second graph, 'Total Production by Gas Cluster using GasProd_MCF', similarly displays production in thousand cubic feet (MCF) by gas cluster. ### Arima Modeling

We opted to utilize the ARIMA (AutoRegressive Integrated Moving Average) model for our analysis due to its effectiveness in capturing and forecasting time series data, which is inherent in our production dataset. ARIMA is well-suited for handling data with trends, seasonality, and autocorrelation, making it a suitable choice for predicting oil and gas production volumes over time. By leveraging the ARIMA model, we aim to exploit the temporal dependencies within the production data to generate accurate forecasts, thereby aiding in strategic decision-making processes. Additionally, ARIMA provides interpretable results, allowing us to gain insights into the underlying patterns driving production fluctuations. Overall, ARIMA offers a robust and versatile framework for modeling time series data, making it a valuable tool for our predictive modeling endeavors.

### ARIMA Model: Oil Clusters

```python
# Extract unique lifecycle stages from the 'Oil_Cluster' column
lifecycle_stages = production['Oil_Cluster'].unique()

# Dictionary to store model summaries for each lifecycle stage
model_summaries = {}

# Iterate over each unique lifecycle stage
for stage in lifecycle_stages:
    # Extract production data for the current stage
    stage_data = production[production['Oil_Cluster'] == stage]['LiquidsProd_BBL']
    
    # Resample the production data to monthly frequency
    stage_data = stage_data.resample('M').mean()
    
    # Drop NaN values
    stage_data.dropna(inplace=True)
    
    # Skip stages with less than 3 data points
    if len(stage_data) < 3:
        continue
    
    # Fit an ARIMA model to the production data for the current stage
    model = ARIMA(stage_data, order=(1,1,1))
    results = model.fit()
    
    # Store the model summary in the dictionary
    model_summaries[stage] = results.summary()
    
    # Calculate RMSE of residuals
    residuals = results.resid
    rmse = np.sqrt(np.mean(residuals**2))
    
    # Pseudo R² calculation
    actual = stage_data
    predictions = results.predict(start=actual.index[0], end=actual.index[-1])
    ss_res = np.sum((actual - predictions) ** 2)
    ss_tot = np.sum((actual - np.mean(actual)) ** 2)
    r_squared = 1 - (ss_res / ss_tot)
    
    # Print model summary, RMSE, and pseudo R² for the current stage
    print(f"{stage} Stage - Model Summary:\n{results.summary()}")
    print(f"{stage} Stage - RMSE: {rmse}")
    print(f"{stage} Stage - Pseudo R²: {r_squared}\n")
```

Additionally, the pseudo R² of 0.2864 suggests that the model may not fully capture the underlying data pattern.

- **Stage 1.0**: This stage exhibits a slightly inferior fit compared to stage 0.0, as indicated by a higher AIC of 2406.50. However, it boasts an impressive pseudo R² of 0.7375, suggesting strong predictive accuracy. The RMSE of 2299.27 is relatively low, indicating a closer alignment between model predictions and actual data compared to other stages.

- **Stage 3.0**: Notably, this stage exhibits a negative pseudo R² of -0.2559, which is unusual and indicative of poor model performance compared to a naive baseline. Moreover, the RMSE is the highest among all stages at 7009.89, further confirming the inadequacy of the model's fit.

- **Stage 4.0**: In this stage, the model shows an AIC of 1935.430, a pseudo R² of 0.1190, and the highest RMSE of 14306.95. These metrics suggest significant discrepancies between the model's predictions and the actual data.

### Exponential Smoothing Model: Oil Clusters

```python
# Convert the index of the production dataframe to datetime format
production.index = pd.to_datetime(production.index)

# Extract unique oil clusters from the 'Oil_Cluster' column and sort them
oil_clusters = sorted(production['Oil_Cluster'].dropna().unique())

# Calculate the number of rows and columns for the subplot grid
num_clusters = len(oil_clusters)
num_cols = min(2, num_clusters)  # Maximum 2 columns
num_rows = (num_clusters - 1) // num_cols + 1

# Create a new figure and subplot grid
fig, axes = plt.subplots(num_rows, num_cols, figsize=(14, 6*num_rows))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Iterate over each unique oil cluster
for i, cluster in enumerate(oil_clusters):
    # Extract oil production data for the current oil cluster
    cluster_data = production[production['Oil_Cluster'] == cluster]['LiquidsProd_BBL'].resample('M').mean().fillna(method='ffill')
    
    # Fit an Exponential Smoothing model to the oil production data for the current oil cluster
    model = ExponentialSmoothing(cluster_data, trend="add", seasonal="add", seasonal_periods=12)
    fit = model.fit()
    
    # Determine the last date in the historical data
    last_date = cluster_data.index[-1]
    
    # Set the forecast start date as the month following the last date in the historical data
    forecast_start_date = last_date + pd.DateOffset(months=1)
    
    # Generate forecast dates for the next 12 months
    forecast_dates = pd.date_range(start=forecast_start_date, periods=12, freq='M')
    
    # Generate forecasts for the next 12 months
    forecast = fit.forecast(12)
    
    # Plot historical data and forecasted values for the current oil cluster
    axes[i].plot(cluster_data.index, cluster_data, label='Historical Data', color='blue')
    axes[i].plot(forecast_dates, forecast, label='Forecast', color='red', linestyle='--')
    axes[i].set_title(f"Exponential Smoothing Forecast for Oil Cluster {cluster}")
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Liquids Production (BBL)')
    axes[i].legend()

# Hide any unused subplots
for j in range(i+1, num_rows*num_cols):
    fig.delaxes(axes[j])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Convert the index of the production dataframe to datetime format
production.index = pd.to_datetime(production.index)

# Extract unique lifecycle stages from the 'Lifecycle_Stage' column and sort them
lifecycle_stages = sorted(production['Lifecycle_Stage'].dropna().unique())

# Calculate the number of rows and columns for the subplot grid
num_clusters = len(lifecycle_stages)
num_cols = min(2, num_clusters)  # Maximum 2 columns
num_rows = (num_clusters - 1) // num_cols + 1

# Create a new figure and subplot grid
fig, axes = plt.subplots(num_rows, num_cols, figsize=(14, 6*num_rows))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Iterate over each unique lifecycle stage
for i, stage in enumerate(lifecycle_stages):
    # Extract production data for the current lifecycle stage
    stage_data = production[production['Lifecycle_Stage'] == stage]['LiquidsProd_BBL'].resample('M').mean().fillna(method='ffill')
    
    # Check if there's enough data to fit the model
    if len(stage_data) < 24:  # Ensuring at least two full seasonal cycles
        print(f"Not enough data to fit model for stage {stage}.")
        continue
    
    # Fit an Exponential Smoothing model to the production data for the current lifecycle stage
    model = ExponentialSmoothing(stage_data, trend="add", seasonal="add", seasonal_periods=12)
    fit = model.fit()
    
    # Determine the last date in the historical data
    last_date = stage_data.index[-1]
    
    # Set the forecast start date as the month following the last date in the historical data
    forecast_start_date = last_date + pd.DateOffset(months=1)
    
    # Generate forecast dates for the next 12 months
    forecast_dates = pd.date_range(start=forecast_start_date, periods=12, freq='M')
    
    # Generate forecasts for the next 12 months
    forecast = fit.forecast(12)
    
    # Plot historical data and forecasted values for the current lifecycle stage
    axes[i].plot(stage_data.index, stage_data, label='Historical Data', color='blue')
    axes[i].plot(forecast_dates, forecast, label='Forecast', color='red', linestyle='--')
    axes[i].set_title(f"Exponential Smoothing Forecast for Lifecycle Stage {stage} for Gas")
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Liquids Production (BBL)')
    axes[i].legend()

# Hide any unused subplots
for j in range(i+1, num_rows*num_cols):
    fig.delaxes(axes[j])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

### STL Modeling

We apply Seasonal-Trend decomposition using LOESS (STL) to analyze the production data of oil clusters across different lifecycle stages. STL is a powerful method for decomposing time series data into its seasonal, trend, and remainder components, enabling us to better understand the underlying patterns and fluctuations within the data. By decomposing the production time series into these components, we gain insights into the seasonal variations, long-term trends, and irregular fluctuations present in the data. This allows us to identify potential patterns, anomalies, or trends that may not be apparent when examining the raw production data alone. Ultimately, applying STL helps us to uncover hidden patterns and better understand the dynamics of oil production across various lifecycle stages, aiding in decision-making and forecasting processes within the oil and gas industry.

### STL Model: Oil Clusters

```python
# Ensure the index is in datetime format
production.index = pd.to_datetime(production.index)

# Extract unique oil clusters
oil_clusters = production['Oil_Cluster'].dropna().unique()

for cluster in oil_clusters:
    print(f"Processing STL for Oil Cluster: {cluster}")
    
    # Extract the time series for the current oil cluster
    cluster_data = production[production['Oil_Cluster'] == cluster]['LiquidsProd_BBL'].resample('M').mean().fillna(method='ffill')
    
    # Ensure there's enough data to perform decomposition
    if len(cluster_data) < 24:
        print(f"Not enough data for STL decomposition for {cluster} cluster.")
        continue
    
    # Apply STL decomposition
    stl = STL(cluster_data, seasonal=13, robust=True)
    result = stl.fit()
    
    # Plotting components
    fig = result.plot()
    plt.subplots_adjust(top=0.9)  # Adjust top spacing
    fig.suptitle(f'STL Decomposition for Oil Cluster {cluster}', fontsize=16)
    plt.tight_layout()  # Adjust layout
    plt.show()

    # Calculate RMSE and R² using the last segment of the data as an example
    # Note: This calculation is conceptual and depends on having a comparable period of actual vs. fitted values
    fitted_values = result.trend + result.seasonal
    actual_values = cluster_data  # The original time series
    
    # Ensure the comparison is valid by aligning lengths; this is a conceptual example
    min_length = min(len(fitted_values), len(actual_values))
    fitted_values = fitted_values[-min_length:]
    actual_values = actual_values[-min_length:]
    
    rmse = mean_squared_error(actual_values, fitted_values, squared=False)
    r_squared = r2_score(actual_values, fitted_values)
    
    print(f"Oil Cluster {cluster} - RMSE: {rmse}, R²: {r_squared}")
```


- **Oil Cluster 0.0**: RMSE is quite high at 1971.4579, and R² is 0.4602, indicating that around 46% of the variance in the data is explained by the model.

- **Oil Cluster 1.0**: Shows a higher RMSE of 2518.8147, but a better R² of 0.6850, suggesting that the model explains approximately 68.5% of the variance.

- **Oil Cluster 2.0**: Here, the RMSE is 1865.2305, and the R² is the highest among the clusters at 0.8818, indicating that the model explains about 88.18% of the variance, which is quite good.

- **Oil Cluster 3.0**: This cluster has the lowest RMSE of 874.9746, and an R² of 0.8012, which is quite robust.

- **Oil Cluster 4.0**: The RMSE increases again to 1743.5339 with an R² of 0.6825, similar to cluster 1.0 in terms of the proportion of variance explained.

```python
# Display headers
display(HTML('<h2>Oil and Gas Production Prediction</h2>'))
display(HTML('<b>Life Cycle Stage Key:</b><br>Old = Years 2009 to 2019<br>Middle = Years 2020 to 2021<br>Young = Years 2022 to 2023'))

# Widget definitions for user interaction
forecast_table_output = widgets.Output()
well_type_dropdown = widgets.Dropdown(options=['OIL', 'GAS'], description='Well Type:', style={'description_width': 'initial'})
lifecycle_stage_dropdown = widgets.Dropdown(options=production['Lifecycle_Stage'].dropna().unique().tolist(), description='Lifecycle Stage:', style={'description_width': 'initial'})
well_name_dropdown = widgets.Dropdown(description='Top 10 Wells:', style={'description_width': 'initial'})
forecast_button = widgets.Button(description='Show Forecast', button_style='info')
clear_button = widgets.Button(description='Clear Selections', button_style='warning')
output = widgets.Output()

def update_well_name_options(*args):
    filtered_df = production[(production['ENVWellType'] == well_type_dropdown.value) & (production['Lifecycle_Stage'] == lifecycle_stage_dropdown.value) & (production['ENVWellStatus'] == 'PRODUCING')]
    column = 'LiquidsProd_BBL' if well_type_dropdown.value == 'OIL' else 'GasProd_MCF'
    top_wells = filtered_df.groupby('WellName')[column].mean().nlargest(10).index.tolist()
    well_name_dropdown.options = top_wells

def on_forecast_button_clicked(b):
    with output:
        output.clear_output()
        if not well_name_dropdown.value:
            print("Please select a well name.")
            return

        selected_well_name = well_name_dropdown.value
        column = 'LiquidsProd_BBL' if well_type_dropdown.value == 'OIL' else 'GasProd_MCF'
        well_data = production[production['WellName'] == selected_well_name][column].resample('M').mean().fillna(method='ffill')
        
        # ARIMA model
        model = ARIMA(well_data, order=(5,1,0))
        try:
            fit_model = model.fit()
        except Exception as e:
            print(f"Model fitting error: {e}")
            return

        forecast = fit_model.forecast(steps=12)
        forecast = forecast.clip(lower=0)  # Ensure non-negative forecasts

        plt.figure(figsize=(12, 8))
        plt.plot(well_data.index, well_data, label='Historical Data', color='blue')
        future_index = pd.date_range(start=well_data.index[-1] + pd.offsets.MonthBegin(1), periods=12, freq='M')
        plt.plot(future_index, forecast, label='Forecasted Production', color='green', linestyle='-.')
        plt.title(f"12-Month Forecast for {selected_well_name}")
        plt.xlabel('Date')
        plt.ylabel(column)
        plt.legend()
        plt.show()
        
        with forecast_table_output:
            forecast_table_output.clear_output()
            forecast_df = pd.DataFrame({'Date': future_index, 'Forecasted Amount': forecast.values})
            display(forecast_df)

def on_clear_button_clicked(b):
    well_type_dropdown.value = well_type_dropdown.options[0]
    lifecycle_stage_dropdown.value = lifecycle_stage_dropdown.options[0]  # Ensure this resets correctly
    well_name_dropdown.options = []
    output.clear_output()
    forecast_table_output.clear_output()
    update_well_name_options()

well_type_dropdown.observe(update_well_name_options, 'value')
lifecycle_stage_dropdown.observe(update_well_name_options, 'value')
forecast_button.on_click(on_forecast_button_clicked)
clear_button.on_click(on_clear_button_clicked)
update_well_name_options()  # Initialize dropdown options

# Layout configuration
box_layout = widgets.Layout(display='flex', flex_flow='column', align_items='stretch', border='solid', width='60%')
widgets_box = widgets.VBox([well_type_dropdown, lifecycle_stage_dropdown, well_name_dropdown, forecast_button, clear_button, output, forecast_table_output], layout=box_layout)

display(widgets_box)
```
